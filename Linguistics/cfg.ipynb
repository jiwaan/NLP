```python
import nltk

grammar = nltk.data.load("grammar.cfg")

sent = 'Let me know if you have any questions on the assignment' # your input sentense

parser = nltk.ChartParser(grammar)    # load the parser

tokens = nltk.tokenize.word_tokenize(sent)

trees = parser.parse(tokens)

out = open("sent3.txt", "w") # writing the parsing result on a file

for i, tree in enumerate(trees):
    # print(tree) # print tree
#     tree.draw() # draw tree
    out.write(f'{i+1}' + str(tree)+"\n\n")

out.close()
```

```python
for i, tree in enumerate(trees):
    print(tree)
    plt.figure(figsize=(10, 10))
    tree.pretty_print(unicodelines=True)
    plt.savefig(f"parse_tree_{i}.png")
```

```python
for tree in trees:
    tree.draw()
```

### sent 5

```python
import nltk

grammar = nltk.data.load("grammar.cfg")

sent = 'Which countries does he want to travel' # your input sentense

parser = nltk.ChartParser(grammar)    # load the parser

tokens = nltk.tokenize.word_tokenize(sent)

trees = parser.parse(tokens)

out = open("sent5.txt", "w") # writing the parsing result on a file

for i, tree in enumerate(trees):
    out.write(f'{i+1}' + str(tree)+"\n\n")

out.close()
```

```python
import matplotlib.pyplot as plt
for i, tree in enumerate(trees):
    print(tree)
    plt.figure(figsize=(10, 10))
    tree.pretty_print(unicodelines=True)
    # plt.savefig(f"parse_tree_{i}.png")
```
